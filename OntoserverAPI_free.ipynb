## Expression Constraint Language to query SNOMED coded data, using Ontoserver's API

Test script to get list of concept ID's which match an ECL query

CSIRO usese Australian version of SNOMED, NHS D has Ontoserver for UK version (requires authorisation)


examples here: https://documenter.getpostman.com/view/634774/TVsuBmc9#f5674478-0c0b-4625-8eaa-4b3896ca794c


```python
#!pip install --user fhir.resources 
# you will probably need to install this
import pandas as pd
import numpy as np
import requests #for Rest calls

#Fhir ValueSet makes parsing the json responses easier.
from fhir.resources.valueset import ValueSet
 
```

```python
#function to go from ECL query -> results
def PopulateSetWithECL(ecl):
    #Compose the API call.
    endpoint= "https://r4.ontoserver.csiro.au/fhir/"
    expandAPI="/ValueSet/$expand"
    sctValueSetUrl='http://snomed.info/sct?fhir_vs=ecl/'    
    urlParam={'url':sctValueSetUrl+ecl}    
    
    #Execute the Get Request, and parse the JSON response to FHIR ValueSet
    response=requests.get(endpoint+expandAPI,params=urlParam)
    j=response.json()
        #add every code to a Set
    _set=set()
    for e in j['expansion']['contains']:
        _set.add(e['code'])
    return _set
```

```python
# example 1

PopulateSetWithECL('<<106048009')
```

```python
# example 2
PopulateSetWithECL('(<< 105969002 AND << 85828009): 116676008 != << 409774005')
```
